```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
data = pd.read_csv('olympicmedal.csv',encoding = "ISO-8859-1")
```

```python
df=data.drop(columns=['City','Discipline', 'Event', 'Athlete', 'Gender','Country'])
```

```python
print(df)
```

```python
#2008 Olympics top 10
#df = pd.read_csv('olympicmedal.csv',encoding = "ISO-8859-1",delimiter=',', skipinitialspace=True, header=None)
df=data.drop(columns=['City','Discipline', 'Event', 'Athlete', 'Gender','Country'])
df['Medal'] = data['Medal'].map({"Bronze": 1, "Silver": 2, "Gold": 3})
df2=df.loc[df['Year']==2008.0]
df41=df2.groupby(['Country_Code'])['Medal'].count().nlargest(10)
df41=df41.reset_index()

print(df41)

sns.set_theme(style="whitegrid")
import matplotlib.pyplot as plt
plt.figure(figsize=(12,4))
sns.barplot(x='Country_Code',y='Medal',data=df41).set_title('2008 Olympics top 10')
```

```python
#particular sport for a country
df34=df.loc[(df['Year']==2008.0) & (df['Sport']=='Aquatics')]
df34.groupby(['Country_Code'])['Medal'].count()
sns.set_theme(style="whitegrid")
import matplotlib.pyplot as plt
plt.figure(figsize=(12,4))
sns.countplot(x='Country_Code', data=df34).set_title('Aquatics 2008 Medal tally')
```

```python
#1976 to 2008 Olympics Men tally 
df.drop(df[df['Event_gender']=='W'].index, inplace=False)
df.groupby(['Country_Code'])['Medal'].count().nlargest(10)
```

```python
#1976 to 2008 USA Men tally: Which sport USA gets most medals?
df=df[df["Country_Code"]=='USA']
dfus=df.drop(df[df['Event_gender']=='W'].index, inplace=False)
dfus.groupby(['Sport'])['Medal'].count().nlargest(10)
result = dfus.groupby(['Sport'])['Medal'].count().nlargest(10).reset_index().sort_values('Medal')
sns.set_theme(style="darkgrid")
plt.figure(figsize=(15,4))
sns.countplot(x='Sport', data=dfus,order=result['Sport']).set_title('1976 to 2008 USA Men tally: Which sport USA gets most medals?')
```

```python
#Predicting number of medals for Aquatics for USA

df=df[(df["Country_Code"]=='USA') & (df["Sport"]=='Aquatics')]
df12=df.groupby(['Year'])['Medal'].count()
df12=df12.reset_index()
sns.lmplot(x="Year", y="Medal",data=df12, markers=["o"])
```

```python
df=df[(df["Country_Code"]=='USA') & (df["Sport"]=='Aquatics')]
df12=df.groupby(['Year'])['Medal'].count()
df12=df12.reset_index()
sns.lmplot(x="Year", y="Medal",data=df12, markers=["o"])
from scipy import stats
slope, intercept, r_value, p_value, std_err = stats.linregress(df12['Year'],df12['Medal'])
print(slope)
print(intercept)
ax = sns.regplot(x="Year", y="Medal", data=df12, color='b',line_kws={'label':"y={0:.5f}x+{1:.5f}".format(slope,intercept)})
ax.legend()
ax.set_title('USA Aquatics Medal prediction', fontdict={'fontsize': 15, 'fontweight': 'medium'})
plt.show()
```

```python
#Checking for competitors part 1
df=data.drop(columns=['City','Discipline', 'Event', 'Athlete', 'Gender','Country'])
df=df[(df["Sport"]=='Wrestling')]
df['Medal'] = data['Medal'].map({"Bronze": 1, "Silver": 2, "Gold": 3})
df12=df.groupby(['Country_Code','Year'])['Medal'].count()
df12=df12.reset_index()
df13=df12[df12['Country_Code'].isin(df12['Country_Code'].value_counts()[df12['Country_Code'].value_counts()>7].index)]

# print(df13)
sns.lmplot(x="Year", y="Medal",data=df13,hue="Country_Code")

```

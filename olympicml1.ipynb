```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
data = pd.read_csv('olympicmedal.csv',encoding = "ISO-8859-1")
data
```

```python
df=data.drop(columns=['City','Discipline', 'Event', 'Athlete', 'Gender','Country','Event_gender'])
df['Medal'] = data['Medal'].map({"Bronze": 1, "Silver": 2, "Gold": 3})
df
```

```python
#particular sport
df1=df.loc[(df['Sport']=='Aquatics')]
df2=df1.groupby(['Year','Country_Code'])['Medal'].count()
```

```python
#Appearing more than 7 Olympics in the past
df2=df2.reset_index()
# df2.shape
df3=df2[df2['Country_Code'].isin(df2['Country_Code'].value_counts()[df2['Country_Code'].value_counts()>7].index)]
df3['Year'].unique()
```

```python
df3['Country_Code'].unique()
```

```python

df4=df3.groupby(['Country_Code','Year']).sum()
df4
```

```python
df4=df4.reset_index()
df5=df4.loc[(df4['Country_Code']=='USA')]
df5
```

```python
reg=linear_model.LinearRegression()
reg.fit(df5[['Year']],df5.Medal)
```

```python
reg.predict([[2020]])
```

```python
# AUS vs USA
df6=df4.loc[(df4['Country_Code']=='AUS')]
df6
```

```python
reg.fit(df6[['Year']],df6.Medal)
```

```python
reg.predict([[2020]])
```

```python
#total medals for USA, CHN, AUS

df10=df.groupby(['Country_Code','Year'])['Medal'].count()
df10=df10.reset_index()
df11=df10.loc[(df10['Country_Code']=='USA')]
df12=df10.loc[(df10['Country_Code']=='CHN')]
df13=df10.loc[(df10['Country_Code']=='AUS')]
```

```python
#Predicted USA 2020 overall medal count
reg.fit(df11[['Year']],df11.Medal)
reg.predict([[2020]])
```

```python
#Predicted CHN 2020 overall medal count
reg.fit(df12[['Year']],df12.Medal)
reg.predict([[2020]])
```
